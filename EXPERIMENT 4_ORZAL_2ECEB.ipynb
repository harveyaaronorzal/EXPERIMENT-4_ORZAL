```python
import pandas as pd

boards = pd.read_excel("board2.xlsx")

Vis = boards.loc [(boards['Hometown'] == 'Visayas') &
                    (boards['Math'] < 70),
                     ['Name','Gender','Track','Math']
                     ]
Vis
```

```python
import pandas as pd

boards = pd.read_excel("board2.xlsx")

Instru = boards.loc [(boards['Track'] == 'Instrumentation') &
                     (boards['Hometown'] == 'Luzon') &
                     (boards['Electronics'] > 70),
                     ['Name','GEAS','Electronics']
                     ]
Instru
```

```python
import pandas as pd

boards = pd.read_excel("board2.xlsx")

boards['Average'] = boards [['Math','Electronics','GEAS', 'Communication']].mean(axis=1)

Mindy = boards.loc [(boards['Gender'] == 'Female') &
                     (boards['Hometown'] == 'Mindanao') &
                     (boards['Average'] >= 55),
                     ['Name','Track','Electronics','Average']]
Mindy
```

2. Create a visualization that shows how the different features contributes to average grade. Does
chosen track in college, gender, or hometown contributes to a higher average score?



```python
import matplotlib.pyplot as plt

track_avg = boards.groupby('Track')['Average'].mean()                          
plt.figure(figsize=(5,5))                                                      
plt.bar(track_avg.index, track_avg.values, color='lightgreen')                      
plt.xlabel('College Track')                                                     
plt.ylabel('Average Grade')                                                      
plt.title('College Track vs Average Score')
```
